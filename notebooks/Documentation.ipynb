# Data Quality: An ELT solution

## Introducción
La calidad de los datos de los clientes constituye un elemento fundamental para la adecuada gestión del riesgo de Lavado de Activos y Financiación del Terrorismo (LAFT). Datos completos, válidos, consistentes y confiables permiten una correcta implementación de las políticas de conocimiento del cliente (KYC), fortalecen los procesos de debida diligencia y contribuyen a una identificación más precisa de la exposición al riesgo.

En el marco conceptual para la evaluación del **Sistema de Administración del Riesgo de LAFT (SARLAFT)**, la **Superintendencia Financiera de Colombia (SFC)** establece la necesidad de evaluar y realizar seguimiento continuo a la calidad de los datos de los clientes de las entidades vigiladas. Estas evaluaciones forman parte del análisis de la gestión operativa que soporta los programas de prevención del riesgo de LAFT.

Adicionalmente, para analizar la evolución de la calidad de los datos y determinar la efectividad de las acciones de mejora implementadas, es indispensable contar con mediciones comparables en el tiempo y entre diferentes instituciones financieras.

La presente solución toma como referencia las buenas prácticas definidas por la Data Management Association (DAMA), organización que establece estándares internacionales para la gestión de datos, incluyendo dimensiones de calidad tales como:
- Completitud
- Validez
- Unicidad
- Consistencia
- Precisión
- Oportunidad

Estas dimensiones se encuentran en armonía con la Circular Básica Jurídica y fortalecen la gestión integral del riesgo de LAFT.

## Problema
A partir del análisis del contexto normativo y operativo, se identificaron las siguientes problemáticas que justifican el desarrollo del proyecto:
- **Ausencia de un modelo estandarizado de calidad de datos**, alineado con las dimensiones reconocidas por DAMA y los lineamientos del SARLAFT.
- **Dificultad para realizar comparaciones consistentes en el tiempo**, lo que limita el análisis de la evolución de la calidad de la información.
- **Riesgo de inconsistencias en la información de los clientes**, afectando la confiabilidad de los procesos de debida diligencia.
- **Limitada evidencia cuantitativa y trazable** para soportar procesos de supervisión y auditoría, tanto internos como externos.

En consecuencia, la organización carece de un mecanismo estructurado, automatizado y repetible que permita:
- Medir objetivamente la calidad de los datos.
- Documentar los resultados de cada evaluación.
- Generar indicadores comparables entre periodos.
- Proporcionar soporte técnico ante requerimientos regulatorios.

El problema no radica únicamente en la posible existencia de errores en los datos, sino en la ausencia de una arquitectura formal que permita gestionarlos estratégicamente.

## Solución
Para abordar las problemáticas identificadas, se propone el diseño e implementación de un sistema automatizado de evaluación de calidad de datos de clientes, basado en una arquitectura ELT (Extract, Load, Transform).

### Objetivos de la solución
- Diseñar un sistema automatizado para evaluar la calidad de los datos de clientes.
- Alinear la medición con los estándares DAMA y el marco del SARLAFT.
- Permitir evaluaciones comparables en el tiempo.
- Facilitar reportes claros, trazables y auditables para entes de control internos y externos.

### Enfoque metodológico: Arquitectura ELT

La solución se fundamenta en una arquitectura ELT, en la cual:

1. **Extract (Extracción)**
Se obtienen las tablas críticas del sistema fuente, garantizando integridad y completitud de la información original.
2. **Load (Carga)**
Los datos extraídos se almacenan en un entorno estructurado que permite su preservación como datos crudos (raw data). Esta etapa asegura trazabilidad y repetibilidad del proceso.
3. **Transform (Transformación)**
Sobre los datos cargados se ejecutan las reglas de evaluación de calidad, calculando métricas asociadas a dimensiones como:
- Completitud
- Validez
- Unicidad
- Consistencia

Los resultados se estructuran bajo un modelo que permite:
- Registro por ejecución.
- Análisis histórico.
- Comparación entre periodos.
- Generación de indicadores cuantitativos.

### Valor estratégico de la solución
La implementación del sistema permite:
- Convertir la calidad de datos en un indicador medible.
- Cumplir con los lineamientos regulatorios de la SFC.
- Fortalecer la evidencia documental ante auditorías.
- Reducir riesgos operativos asociados a información inconsistente.
- Permitir seguimiento continuo y mejora progresiva de la calidad.
